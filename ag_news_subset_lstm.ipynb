```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

dataset, info = tfds.load('ag_news_subset', with_info=True, as_supervised=True)

train_data, test_data = dataset['train'], dataset['test' ]

training_sentences = []
testing_sentences = []

# # add sentences to lists with for loop
# # training_sentences and labels
# for ...

# # testing_sentences and labels
# for .....

# # preprocess text data

# # build a model
```

```python
training_labels = []
test_labels = []
```

```python
# # add sentences to lists with for loop
# # training_sentences and labels
for i, j in train_data:
  training_sentences.append(i.numpy().decode('utf8'))
  training_labels.append(j.numpy())
  
```

```python
for i, j in train_data:
  testing_sentences.append(i.numpy())
  test_labels.append(j.numpy())
```

```python
training_labels = np.array(training_labels)
test_labels = np.array(test_labels) 
```

```python
VOCAB_SIZE = 10000
OOV_TOKEN = "<oov>"
MAX_LEN = 100

tokenizer = Tokenizer(num_words = VOCAB_SIZE, oov_token = "<oov>")
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index

sequences = tokenizer.texts_to_sequences(training_sentences)
padded = pad_sequences(sequences, maxlen=MAX_LEN, padding='post')
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense,Bidirectional
```

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(VOCAB_SIZE, 64, input_length = MAX_LEN)),
model.add(Bidirectional(LSTM(64, return_sequences=True))),
model.add(Bidirectional(LSTM(64))),

model.add(tf.keras.layers.Flatten()),
model.add(tf.keras.layers.Dense(64, activation='relu')),
model.add(tf.keras.layers.Dense(4, activation='softmax')) 
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='nadam', metrics='accuracy')
```

```python
model.fit(padded, training_labels, epochs=5)
```
